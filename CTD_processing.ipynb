```python
import os
import numpy as np
import datetime
import matplotlib.pyplot as plt
import pickle
```

```python
roots = 'C:\\Users\\simcosta\\Documents\\doc\\Campo_2023\\Campo_2023\\dados_brutos\\CTD_jfe\\'
```

```python
ctdfilenames = os.listdir(roots)
dados_ctd = []
time = []
for i in range(len(ctdfilenames)):
    with open(roots+ctdfilenames[i]) as file:
        time = []
        press = []
        temp = []
        sal = []
        chla = []
        turb = []
        oxy = []
        oxy_sat = []
        dens = []
        for line in file:
            if line[0:10] == '2023/11/28':
                ano = int(line[0:4])
                mes = int(line[5:7])
                dia = int(line[8:10])
                hora = int(line[11:13])
                minuto = int(line[14:16])
                seg = int(line[17:19])
                trasht = datetime.datetime.timestamp(datetime.datetime(ano, mes, dia, hora, minuto, seg))
                
                line_break = line.split(',')
                depth_teste = ((float(line_break[1])*10000)/(float(line_break[6])*9.79499))
                if int(depth_teste) >= 0.1: 
                    time.append(trasht)
                    press.append(((float(line_break[1])*10000)/(float(line_break[6])*9.79499)))
                    temp.append(float(line_break[2]))
                    sal.append(float(line_break[3]))
                    chla.append(float(line_break[9]))
                    turb.append(float(line_break[10]))
                    oxy.append(float(line_break[11]))
                    dens.append(float(line_break[6]))
            
        
        dados_ctd.append([time, press, temp, sal, chla, turb, oxy, dens])
            
```
